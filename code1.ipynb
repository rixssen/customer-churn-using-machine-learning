```python
import pandas as pd
```

```python
df = pd.read_csv("customer_churn_data.csv")
```

```python
df
```

```python
y = df[["Churn"]]
x = df[["Age" , "Gender" , "MonthlyCharges" , "Tenure"]]
```

```python
y["Churn"] = y["Churn"].apply( lambda x:1 if x == "Yes" else 0)
```

```python
y["Churn"].value_counts()
```

```python
y
```

```python
x["Gender"] = x["Gender"].apply(lambda x:1 if x == "Female" else 0)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x , y , test_size= 0.2)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
scaled_x_test = scaler.fit_transform(x_test)
```

```python
scaled_x_test
```

```python
x_train
```

```python
from sklearn.metrics import accuracy_score
def modelperformance(predictions):
    print("Accuracy score on model is {}".format(accuracy_score(y_test , predictions)))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
```

```python
log_model.fit(x_train , y_train)
```

```python
y_pred = log_model.predict(x_test)
```

```python
modelperformance(y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
param_grid = {
    "n_neighbors" : [3,5,7,9],
    "weights" : ["uniform","distance"], 


}
```

```python
gridkn = GridSearchCV(KNeighborsClassifier(),param_grid , cv = 5)
```

```python
gridkn.fit(x_train,y_train)
```

```python
gridkn.best_params_
```

```python
y_pred = gridkn.predict(x_test)
```

```python
modelperformance(y_pred)
```

```python
from sklearn.svm import SVC
```

```python
param_grid = {
    "C" : [0.01,0.1,0.5,1],
    "kernel" : ["linear","rbf", "poly"],
    
}
```

```python
svc = SVC()
```

```python
gridsvc= GridSearchCV(svc , param_grid , cv= 5)
```

```python
gridsvc.fit(x_train,y_train)
```

```python
y_pred= gridsvc.predict(x_test)
```

```python
modelperformance(y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
param_grid = {
    "criterion" : ["gini", "entropy"],
    "splitter"  : ["best", "random"],
    "max_depth" : [ None,10, 20, 30],
    "min_samples_split" : [2,5,10],
    "min_samples_leaf" : [1,2,4],
            }
```

```python
grid_tress = GridSearchCV(DecisionTreeClassifier(), param_grid , cv=5)
```

```python
grid_tress.fit(x_train,y_train)
```

```python
y_pred = grid_tress.predict(x_test)
```

```python
modelperformance(y_pred)
```
